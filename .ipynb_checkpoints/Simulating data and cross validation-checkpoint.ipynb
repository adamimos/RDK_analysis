```python
def p_sigmoid(y0,a,T,b,c):
    return y0 + a /(1+np.exp((c-T)/b))



def simulate_data(num_trials, y0, a, t, b, num_bins):
    #let's generate some data and then see how well we can fit it with MCMC


    coherence_bins = np.array(np.linspace(-1, 1, num = num_bins))

    coherences = np.random.choice(coherence_bins, size = num_trials)


    observations = np.random.binomial(1, p_sigmoid(y0, a, t, b, coherences))
    
    return observations, coherences, coherence_bins


observations, coherences, coherence_bins = simulate_data(num_trials = 600, y0 = .1, a = .8, t = 0, b = .2, num_bins = 10)
```

```python
def calc_psych_plot(observations, coherences, coherence_bins):

    means = np.array([observations[np.digitize(coherences, coherence_bins + .1) == i].mean() 
                      for i in np.arange(len(coherence_bins))])
    ns = np.array([len(observations[np.digitize(coherences, coherence_bins + .1) == i]) 
                 for i in np.arange(len(coherence_bins))])
    
    std = np.sqrt(means*(1-means) / ns)
    
    return means, std

means, std = calc_psych_plot(observations, coherences, coherence_bins)
```

```python

ax = plt.subplot(111)
plt.plot(coherence_bins, p_sigmoid(.1, .8, 0, .2, coherence_bins), c = 'r', linewidth = 3)
plt.errorbar(coherence_bins,*calc_psych_plot(observations, coherences, coherence_bins), c = 'k', fmt = 'o')

plt.title('psychophysical curve')
simpleaxis(ax)
```

```python
import pymc3 as pm
import theano.tensor as tt


def simple_psych_model(coherences, observations):

    with pm.Model() as simple_model:
        y0_ = pm.Uniform("y0",0,1)
        a_ = pm.Uniform("a",0,1-y0_)
        T_ = pm.Uniform("T",-1,1)
        b_ = pm.Uniform("b",-1,1)

        theta_ = y0_ + a_ * tt.nnet.nnet.sigmoid( - (coherences-T_)/b_)    
        theta = pm.Deterministic("theta",theta_)

        NR = pm.Bernoulli("NR",theta,observed=observations)
        
    return simple_model


simple_psych_model(coherences, observations)

with simple_model:
    step = pm.Metropolis()
    simple_trace = pm.sample(10000, tune=5000,step=step)
```

```python
_ = pm.traceplot(simple_trace, lines = {"y0": .1, "a": .8, "T": 0, "b": .2} )
```

```python
T_traces = simple_trace['T']
a_traces = simple_trace['a']
y0_traces = simple_trace['y0']
b_traces = simple_trace['b']

plt.figure(figsize = [5, 10])
c = np.linspace(-1,1)
ax = plt.subplot(311)

psych_ = lambda y0,a,T,b,c: y0 + a /(1+np.exp((c-T)/b))

for i in range(8000,9999):
    plt.plot(c, p_sigmoid(y0_traces[i],a_traces[i],T_traces[i], -b_traces[i],c),alpha=0.01,c='m')


plt.plot(c, p_sigmoid(.1, .8, 0, .2, c), c = 'k', linewidth = 3)
plt.errorbar(coherence_bins, *calc_psych_plot(observations, coherences, coherence_bins), c = 'k', fmt = 'o')
    
    
plt.legend(loc="upper left")
plt.title(r"""Posterior distributions of the variables
    $\lambda_1,\;\lambda_2,\;\tau$""")
#plt.xlim([-1,1 ])
plt.xlabel("$\lambda_1$ value")

simpleaxis(ax)
```

```python
def Switching_Model(observations, coherences, num_trials):
    time = np.arange(0, 2*num_trials)
    with pm.Model() as switching_model:

        #model before 1st switch

        y0 = pm.Uniform("y0_0",0,1)
        a = pm.Uniform("a_0",0,1 - y0)
        T_0 = pm.Uniform("T_0",-1,1)
        b = pm.Uniform("b_0",-10,10)

        #model after 1st switch


        y0_2 = pm.Uniform("y0_2",0,1)
        a_2 = pm.Uniform("a_2",0,1 - y0_2)
        T_2 = pm.Uniform("T_2",-1,1)
        b_2 = pm.Uniform("b_2",-10,10)



        s1 = pm.DiscreteUniform('switchpoint1',400, 800, testval = 600)

        tau_y1 = pm.math.switch(time >= s1, y0_2, y0)    
        tau_a1 = pm.math.switch(time >= s1, a_2, a) 
        tau_T1 = pm.math.switch(time >= s1, T_2, T_0) 
        tau_b1 = pm.math.switch(time >= s1, b_2, b) 


        theta_ = y0 + a * tt.nnet.nnet.sigmoid(-(coherences-tau_T1)/tau_b1)    
        theta = pm.Deterministic("theta",theta_)


        NR = pm.Bernoulli("NR",theta,observed=observations)
        
    return switching_model

switching_model = Switching_Model(observations, coherences, 600)

with switching_model:
    step = pm.Metropolis()
    switching_trace = pm.sample(1000000, tune=5000,step=step)
```

```python
from scipy.stats import norm

#now let's generate data from a time-evolving threshold

num_trials = 1000

time = np.arange(num_trials)

threshold = .4 - 100*norm.pdf(time, loc = 200, scale = 50)

plt.plot(threshold)
```

```python
y01 = .1
a1 = .9
b1 = .1

psych = lambda coherences, threshold: y01 + (a1/(1+np.exp(-(coherences-threshold)/b1)))

coherences = np.random.choice(coherence_bins, size = num_trials)
observations = np.random.binomial(1, psych(coherences, threshold))
```

```python
means = np.array([observations[np.digitize(coherences, coherence_bins + .1) == i].mean() 
                  for i in np.arange(len(coherence_bins))])
ns = np.array([len(observations[np.digitize(coherences, coherence_bins + .1) == i]) 
             for i in np.arange(len(coherence_bins))])

std = np.sqrt(means*(1-means) / ns)



ax = plt.subplot(111)

for i in range(num_trials):
    plt.plot(c, psych(c, threshold[i]), c = 'r', alpha = .01)
#plt.plot(coherence_bins, psych(coherence_bins), c = 'r', linewidth = 3)
plt.errorbar(coherence_bins, means, yerr=std, c = 'k', fmt = 'o')

plt.title('psychophysical curve')
simpleaxis(ax)
```

```python
with pm.Model() as GP_model:
    y0 = pm.Uniform("y0",0,1)
    a = pm.Uniform("a",0,1-y0)
    b = pm.Uniform("b",-10,10)
       
    
    #for gaussian process
    l = pm.Uniform('l', 50, 300)
    #l = 100
    
    # uninformative prior on the function variance
    log_s2_f = pm.Uniform('log_s2_f', lower=-5, upper=5)
    s2_f = pm.Deterministic('s2_f', tt.exp(log_s2_f))

    # uninformative prior on the noise variance
    log_s2_n = pm.Uniform('log_s2_n', lower=-10, upper=5)
    s2_n = pm.Deterministic('s2_n', tt.exp(log_s2_n))

    # covariance functions for the function f and the noise
    f_cov =  s2_f* pm.gp.cov.ExpQuad(1, l)

    
    g_p = pm.gp.Latent(cov_func=f_cov)
    
    f = g_p.prior("f", X = time[:, None] )
    
    thresh = pm.Normal("thresh", mu = f, sd = s2_n, shape = num_trials)
       
    theta_ = y0 + a * tt.nnet.nnet.sigmoid(-(coherences-thresh)/b) 
    theta = pm.Deterministic("theta",theta_)
    NR = pm.Bernoulli("NR",theta,observed=observations)
    
with GP_model:
    step = pm.Metropolis()
    GP_trace = pm.sample(50000, tune=1000,step=step)
    
```

```python
#T_traces = trace['T']
a_traces = GP_trace['a']
y0_traces = GP_trace['y0']
b_traces = GP_trace['b']
thresh_traces = GP_trace['thresh']

figsize(15, 10)
c = np.linspace(-1,1)
ax = plt.subplot(221)

#psych = lambda y0,a,T,b,c: y0 + a /(1+np.exp((c-T)/b))

for i in range(29900, 30000):
    plt.plot(thresh_traces[i],alpha=0.005,c='m')

plt.legend(loc="upper left")
plt.title(r"""Posterior distributions of the variables
    $\lambda_1,\;\lambda_2,\;\tau$""")
#plt.xlim([-1,1 ])
plt.xlabel("$\lambda_1$ value")


simpleaxis(ax)



figsize(15, 10)
c = np.linspace(-1,1)
ax = plt.subplot(222)

psych1 = lambda y0,a,T,b,c: y0 + a /(1+np.exp((c-T)/b))

for i in range(29900, 30000):
    for j in np.arange(0, num_trials, 5):
        plt.plot(c,psych1(y0_traces[i],a_traces[i],thresh_traces[i, j],b_traces[i],c),alpha=0.005,c='m')

        
plt.errorbar(coherence_bins, means, yerr=std, c = 'k', fmt = 'o')

plt.legend(loc="upper left")
plt.title(r"""Posterior distributions of the variables
    $\lambda_1,\;\lambda_2,\;\tau$""")
#plt.xlim([-1,1 ])
plt.xlabel("$\lambda_1$ value")


simpleaxis(ax)
```
